```python
import pandas as pd
```

```python
df_emp = pd.read_csv("emp.csv")
df_dept = pd.read_csv("dept.csv")
```

```python
# Similar to join in SQL, we have merge function here in pandas ( INNER JOIN example )
# SQL equivalent : SELECT * from emp as e  inner join dept as d on e.deptno = d.deptno 
df_inner =df_emp.merge(df_dept,on='deptno',how='inner')
```

```python
df_inner
```

```python
# Similar to join in SQL, we have merge function here in pandas ( LEFT JOIN example )
# SQL equivalent : SELECT * from emp as e  left join dept as d on e.deptno = d.deptno 
df_left =df_emp.merge(df_dept,on='deptno',how='left')
```

```python
df_left
```

```python
# Similar to join in SQL, we have merge function here in pandas ( RIGHT JOIN example )
# SQL equivalent : SELECT * from emp as e  right join dept as d on e.deptno = d.deptno 
df_right =df_emp.merge(df_dept,on='deptno',how='right')
```

```python
df_right
```

```python
# Similar to join in SQL, we have merge function here in pandas ( FULL OUTER JOIN example )
# SQL equivalent ( Oracle )  : SELECT * from emp as e  full outer join dept as d on e.deptno = d.deptno 
df_full_outer =df_emp.merge(df_dept,on='deptno',how='outer')
```

```python
df_full_outer
```

```python
# Similar to join in SQL, we have merge function here in pandas ( INNER JOIN example ) with joining column with different name
# SQL equivalent ( Oracle )  : SELECT * from emp as e  full outer join dept as d on e.deptno = d.dno
df =df_emp.merge(df_dept,left_on='deptno',right_on ='dno',how='inner')
```

```python
# Similar to join in SQL, we have merge function here in pandas ( INNER JOIN example ) with joining column with different name
# SQL equivalent ( Oracle )  : SELECT * from emp as e  full outer join dept as d on e.deptno = d.dno
df =df_emp.merge(df_dept,left_on='deptno',right_on ='dno',how='left')
```
